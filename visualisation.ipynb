```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(style="ticks")

# Initialize the figure with a logarithmic x axis
f, ax = plt.subplots(figsize=(20, 16))
ax.set_xscale("log")

# Load the example planets datase

# Plot the orbital period with horizontal boxes
sns.boxplot(
    df, x="tg", y="price_personal", hue="tg",
    whis=[0, 100], width=.6, palette="vlag"
)

# Add in points to show each observation
sns.stripplot(df, x="tg", y="price_personal", size=4, color=".3")

# Tweak the visual presentation
ax.xaxis.grid(True)
sns.despine(trim=True, left=True)
```

```python
import seaborn as sns
sns.set_theme(style="ticks", palette="pastel")

# Load the example tips dataset


# Draw a nested boxplot to show bills by day and time
sns.boxplot(x="tg", y="price_personal",
            hue="competitor", palette=["m", "g"],
            data=df[df['price_personal'] <= 100000])
sns.despine(offset=10, trim=True)
plt.xticks(rotation=45, 
        ha='right',
        );
```

```python
plt.figure(figsize=(24,15), dpi= 800)
fig, axes = joypy.joyplot(df, column=['price_personal'], by = 'tg', ylim='competitor', figsize=(10,5), hist= True, bins = df['price_personal'].count() // 10,)

# Decoration
plt.title('Распределение изделий по цене Sokolov', fontsize=16)
plt.show()
```

```python
plt.figure(figsize=(16,10), dpi= 80)
fig, axes = joypy.joyplot(df[df['price_personal'] <= 100000], column=['price_personal'], by = 'tg', ylim='competitor', figsize=(10,5), 
                        fade = True,
                        colormap = cm.Pastel1)

# Decoration
plt.title('Распределение изделий конкурентов', fontsize=16)
plt.show()
```
