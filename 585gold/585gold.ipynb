```python
psize = 113
array = range(1,psize)
new_prices = list()
old_prices = list()
names = list()
links = list()
competitors = list()
discounts = list()
data = list()
prx = {
    'http' : '45.11.124.127:8000',
    'https' : '45.11.124.127:8000'
}
for p in tqdm(array):
    url = f'https://zoloto585.ru/catalog/yuvelirnye_izdeliya/dop_44_dk_pdk/?page={p}'
    try:
        resp = requests.get(url,headers=headers,cookies=cookies,proxies=prx)
    except:
        time.sleep(60)
        resp = requests.get(url,headers=headers,cookies=cookies,proxies=prx)
    time.sleep(1)
    soup  = bs(resp.content, 'html.parser')
    fff = soup.find_all('div', class_='product-card__main-data')
    for art in fff:
        newp = art.find('span', class_='product-card__total-price').text[:-2]
        op = art.find('span', class_='product-card__old-price').text[:-2]
        name = art.find('h3',attrs={'class':'product-card__title'}).text
        link = art.find('a',attrs={'class':'product-card__link'})
        try:
            disc40 = art.find('div',class_='product-card__coupon').text
        except:
            disc40 = np.nan
        link = link['href']
        competitor = "585 GOLD"
        data.append(datetime.date.today())
        new_prices.append(newp)
        old_prices.append(op)
        names.append(name)
        links.append(link)
        competitors.append(competitor)
        discounts.append(disc40)
            
mylinks = pd.DataFrame({
    "data": data,
    "competitor": competitors,
    "name": names,
    "price": new_prices,
    "price_old": old_prices,
    "arts": links
})
```

```python
mylinks['discount'] = discounts
```

```python
values = list()
inserts = list()
metall = list()
proba = list()
design = list()
pusto = list()
inserts_weights = list()
for link in tqdm(links):
    stones = False
    values.append(np.nan)
    inserts.append(np.nan)
    metall.append(np.nan)
    proba.append(np.nan)
    design.append(np.nan)
    pusto.append(np.nan)
    inserts_weights.append(0)
    resp = requests.get(f'https://zoloto585.ru{link}',headers=headers,cookies=cookies,proxies=prx)
    soup = bs(resp.content,'html.parser')
    fn = soup.find_all('div',attrs={'class':'product-page-specifications__item-name'})
    fv = soup.find_all('div',attrs={'class':'product-page-specifications__item-value'})
    for i in range(len(fn)):
        if "вес" in fn[i].text.lower():
            values[-1] = fv[i].text
            stones = False
        elif "металл" in fn[i].text.lower():
            metall[-1] = fv[i].text
            stones = False
        elif "проба" in fn[i].text.lower():
            proba[-1] = fv[i].text
            stones = False
        elif "дизайн" in fn[i].text.lower():
            design[-1] = fv[i].text
            stones = False
        elif "пустотел" in fn[i].text.lower():
            pusto[-1] = fv[i].text
            stones = False
        if "артикул" in fn[i].text.lower():
            stones = False
        if stones:
            inserts[-1] = fn[i].text.lower()
            prom = fv[i].text
            promcount = prom[:prom.find(" ")]
            count = int(promcount.strip())
            carat = float(prom[prom.find("-")+2:prom.find("кар")-1])
            inserts_weights[-1]  = count * carat * 0.2 + inserts_weights[-1]
        if "камни" in fn[i].text.lower():
            stones = True
mylinks['proba'] = proba
mylinks['value'] = values
mylinks['insert'] = inserts
mylinks['metall'] = metall
mylinks['design'] = design
mylinks['pusto'] = pusto
mylinks['insert_value'] = inserts_weights
mylinks.to_excel('585_disc.xlsx')
```
