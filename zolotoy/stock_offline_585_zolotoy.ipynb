```python
import datetime as dt
import pandas as pd
import numpy as np
import pymssql
import warnings
import calendar
import locale
import glob
import os

warnings.filterwarnings('ignore')
pd.set_option('use_inf_as_na', True)
pd.options.display.max_columns = None
pd.set_option('float_format', '{:.2f}'.format)
locale.setlocale(locale.LC_ALL, 'ru_RU.UTF-8')
```

```python
# SQL-скрипт ниже выгружает данные по остаткам с БД Staging, доступ к ней выдаётся по учётной записи Windows по согласованию

conn = pymssql.connect(server='AX-SQL', database='Staging')
cursor = conn.cursor()

sql = f'''
SELECT [Код склада], [Название склада], [Цена Закупки за 1 шт. (упр сс с НДС за единицу)], [Группа цен],
[Товарное направление], [Товарная группа], [Проба], [Цена Розн., за шт],  [Запрет скидки],
[Код товара],  [Артикул поставщика], [Чистый вес], [ШК], [Общий вес], [Скидка на изделии], [Тип изделия 1]
from [Staging].[Reports].[Remainings_8h]
where [Товарное направление] in (N'БК', N'ЦБ')
'''

sql_query = pd.read_sql_query(sql, conn)
df = pd.DataFrame(sql_query)
```

```python
# смотрим, прочитался ли актуальный файл
if f'{path}\\МОТИВАЦИОННЫЙ выполнение плана продаж {name_month} {year} - {day}.xlsx' == str(latest_file):
    print('прочитан актуальный файл')
else:
    print(''' 
нужный на сегодня файл отсутствует

ответственный за загрузку: Илларионов Иван
загружает этот файл после формирования своего ежедневного отчёта примерно в 13:00''')
```

```python
sklad = pd.read_excel(latest_file, sheet_name = 'Продажи по ассортименту', header = 2)
sklad = sklad[['Unnamed: 0','Торговая марка', 'Юр лицо', 'Код склада', 'Дата открытия объекта', 'Код объекта', 'Город', 'Наименование объекта']]
sklad1 = sklad[['Код склада', 'Код объекта']].drop_duplicates().reset_index(drop=True)
sklad1 = sklad1[sklad1['Код объекта'].notna()]
```

```python
# отбираем кбк без франчайзи(у них кбк начинается с "8")
sklad1['Код объекта'] = sklad1['Код объекта'].astype(str)
sklad1['8'] = sklad1['Код объекта'].apply(lambda x: x[0])
sklad1 = sklad1[sklad1['8'] != '8']
```

```python
df = df.merge(sklad1, left_on='Код склада', right_on = 'Код склада', how='inner')
```

```python
# можно использовать для расчета чистый вес или общий вес. Чистый вес и общий вес должны совпадать
df['Бирка за грамм'] = df['Цена Розн., за шт'] / df['Чистый вес']
```

```python
df['Скидка на изделии'] = df['Скидка на изделии'].astype(float)
df['Скидка на изделии'] = (df['Скидка на изделии'] / 100).round(2)
```

```python
# Скидка по ПЛ(программе лояльности) не постоянная, она постоянно меняется. 
# Всегда нужно уточнять актуальный промокод у Николая или у Тани
discount = 0.8

df['Скидка_по_ПЛ'] = np.where(
    df['Запрет скидки']=='Да', 1, 
    np.where(
        (df['Запрет скидки']=='Нет') & ((df['Товарное направление']=='ДК')|
                                        (df['Товарное направление']=='ПДК')|
                                        (df['Товарное направление']=='БК')|
                                        (df['Товарное направление']=='ЦБ')|
                                        (df['Товарное направление']=='ИФ')|
                                        (df['Товарное направление']=='СИ')|
                                        (df['Товарное направление']=='ДФРС')|
                                        (df['Товарное направление']=='ВЫРИЦА')|
                                        (df['Товарное направление']=='ЧАСЫ')), discount, 1
        )
    )
```

```python
df['Запрет скидки'].value_counts()
```

```python
df['Цена после ПЛ'] = np.where(
    df['Запрет скидки'] == 'Нет', df['Цена Розн., за шт'] * df['Скидка_по_ПЛ'], df['Цена Розн., за шт']
    )
```

```python
df['Цена за грамм после ПЛ'] = df['Цена после ПЛ'] / df['Чистый вес']
```

```python
result = df.groupby(by = 'Товарная группа', as_index = False)[['Цена после ПЛ', 'Чистый вес', 'Цена Закупки за 1 шт. (упр сс с НДС за единицу)']].agg('sum')
```

```python
result = result[result['Товарная группа'].str.contains('БК КОЛЬЦА|БК КОЛЬЦА ОБРУЧ|БК ПОДВЕС ДЕКОР|БК ПОДВЕС КУЛЬТ|БК СЕРЬГИ|ЦБ БРАСЛЕТЫ|ЦБ ЦЕПИ')]
```

```python
result
```

```python
result['Наценка остатки оффлайн'] = result['Цена после ПЛ'] * 0.8 / result['Цена Закупки за 1 шт. (упр сс с НДС за единицу)'] - 1
```

```python
result['10%+20%'] = round(result['Цена после ПЛ'] * 0.8 * 0.9 / result['Чистый вес'], 1).astype('int')
```

```python
result['1=2'] = round(result['Цена после ПЛ'] / 0.8 * 0.9 * 0.5 / result['Чистый вес'], 1).astype('int')
```

```python
result.columns = ['tg', 'price', 'weight', 'cost_price', 'margin', 'with_discount_1', 'with_discount_2']
```

```python
stock_price_pl =  ((result['price']).sum() / 0.8 * 0.9 * 0.5 / result['weight'].sum()).round(0)
print('Средняя цена остатков офлайн ПЛ: ', stock_price_pl)
```
